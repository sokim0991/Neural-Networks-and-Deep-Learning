```
import numpy as np
a = np.array([1,2,3,4])
print(a)
```

```
#vectorized version

import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()

#The vectorize version took 1.5 milliseconds. 
print(c)
print("vectorized version:"+str(1000*(toc-tic))+"ms")
```

```
#For loop version:
c = 0

tic = time.time()
for i in range(1000000):
  c += a[i]*b[i]
toc = time.time()

#vectorized version runs over 300 times faster
print(c)
print("For loop:" + str(1000*(toc-tic))+"ms")
```
