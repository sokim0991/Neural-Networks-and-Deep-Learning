```
import numpy as np

A = np.array([[56.0, 0.0, 4.4, 68.0],
             [1.2, 104.0, 52.0, 8.0],
             [1.8, 135.0, 99.0, 0.9]])

print(A)
```

```
#axis = 0: sum virtically
#axis = 1: sum horizontally
cal = A.sum(axis=0)
print(cal)
```

```
#divide matrix A by 1X4 matrix
#reshape module is redundant here
#56.0/59 = 94..
percentage = 100*A/cal.reshape(1,4)
print(percentage)
```
